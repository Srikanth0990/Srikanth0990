```python
def ask_ai(question):

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": question
            }
        ]
    )

    answer = response.choices[0].message.content

    return answer
```

```python
print(ask_ai("Who are you?"))
```

```python
while True:

    question = input("You : ")

    if question.lower() == "exit":
        print("Chat ended.")
        break

    answer = ask_ai(question)

    print("\nAI :", answer)
    print()
```
