```python
!pip install -q groq PyPDF2

import PyPDF2
from groq import Groq

# ==========================================
# 1. CONFIGURATION
# ==========================================
# Replace with your actual Groq API Key
GROQ_API_KEY = "your_groq_api_key_here"

# Path to your PDF file uploaded in Colab
BOOK_FILE = "i-too-had-a-luv-story.pdf"


# ==========================================
# 2. INITIALIZATION
# ==========================================
client = Groq(api_key=GROQ_API_KEY)


def read_pdf(file_path):
    """Extracts text safely from all pages of the PDF."""
    text = ""
    with open(file_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text


# Load PDF content
try:
    book_content = read_pdf(BOOK_FILE)
    print("✅ PDF Loaded Successfully!")
    print("Type 'exit' to quit.\n")
except Exception as e:
    print(f"❌ Failed to load PDF: {e}")
    book_content = ""


# ==========================================
# 3. INTERACTIVE CHAT LOOP
# ==========================================
if book_content:
    while True:
        question = input("You: ")

        if question.strip().lower() == "exit":
            print("Bot: Goodbye!")
            break

        # Construct prompt restricting context to PDF content
        prompt = f"""
You are a helpful PDF assistant.

Answer ONLY using the provided PDF content below.
If the answer is not present in the PDF context, reply EXACTLY with:
Not in PDF

PDF CONTENT:
{book_content[:4000]}

QUESTION:
{question}
"""

        try:
            response = client.chat.completions.create(
                model="llama-3.1-8b-instant",
                messages=[
                    {
                        "role": "user",
                        "content": prompt
                    }
                ],
                temperature=0.2,
                max_tokens=200
            )

            answer = response.choices[0].message.content
            print(f"\nBot: {answer}\n")

        except Exception as e:
            print(f"\nError processing request: {e}\n")
```
