```python
def search_jobs(keyword):

    keyword = keyword.lower()

    results = df[df.astype(str).apply(
        lambda x: x.str.lower().str.contains(keyword)
    ).any(axis=1)]

    return results
```

```python
def naukri_ai(question):

    jobs = search_jobs(question)

    if jobs.empty:
        return "No matching jobs found."

    jobs_text = jobs.head(10).to_string(index=False)

    response = client.chat.completions.create(

        model="llama-3.3-70b-versatile",

        messages=[

            {
                "role":"system",
                "content":"You are an AI Job Recommendation Assistant."
            },

            {
                "role":"user",
                "content":f"""

                The following jobs were found:

                {jobs_text}

                Explain and recommend the most suitable jobs.
                """
            }

        ],

        temperature=0.5,
        max_tokens=1000

    )

    return response.choices[0].message.content
```

```python
while True:

    question = input("Ask Naukri AI : ")

    if question.lower() == "exit":
        break

    print(naukri_ai(question))
```
